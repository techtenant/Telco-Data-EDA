```python
import pandas as pd
import numpy as np
```

```python
#Read Data --> Set from CSV
df = pd.read_csv("telecom_churn.csv")
#First method Head --> Shows First 5 columns of dataset
df.head()
```

```python
#Property --> Shape (Gives you dimensions of the datasets. rows aand columns)
df.shape
```

```python
#Knowing columns -->Get complete cols in dataframe
df.columns
```

```python
#Full Dataset Summary
df.info()
```

```python
#astype method --> to change datatype in this case for churn 
df['churn'] = df['churn'].astype('int64')
df.head()
```

```python
# Statistical summary of data more so numerical feature
df.describe()
```

```python
#Statistical Summary of dataset of non numericals features
df.describe(include=['object'])
```

```python
#Series --> One dimentional array 
# --> making use of some characters seeing which is more dorminant
s = pd.Series(['a','b','c','d','e','a','a','b','b','a','c','e','e'])
print(s.describe())
```

```python
#value_counts method --> Mostly used for boolean to count unique values
# Users stopped using services
df['churn'].value_counts()
```

```python
#Sorting --> sorts values by along either axis
df.sort_values(by ='total day minutes').head()
```

```python
#Sorting Descending --> sorts values by along either axis
df.sort_values(by ='total eve calls', ascending= False).head()
```

```python
#Churn users on international plan

df[(df['churn']==0) & (df['international plan']=='yes')]
```

```python
#How much time (on average) left users spend on phone during dayTime?
df[df['churn']==1]['total day minutes'].mean()
```

```python
#Only users using international plan
df[(df['churn']==0) & (df['international plan']=='yes')]
```

# loc -->Location (Purely label-location based indexer for selection by label.)
#iloc --> Index Location

```python
#specify 1st no. of rows to fetch and no. of cols to fetch
df.loc[0:5, 'state': 'area code']
#We've sliced data but including the 5th
```

```python
df.iloc[0:5,0:3]
#Slices and does not need col names it uses indexes
```

```python
#Getting information of the last cell
df[-1:]
```

```python
#Apply() --> Applies function along input axis/ column of DataFrame.
#Like getting the max in each column
df.apply(np.max)
```

```python
# Decided to use Lambda func expression to get info of those states starting with letter W

df[df['state'].apply(lambda state: state[0]=='W')].head()
```

```python
#Data Mapping concept like converting the values of the column to another value
# i.e --> Yes - True no - False etc

#Easily works with dictonary --> Key value pair 
#Will use the international plan column meaning my key will be no and yes whereas value being false and true respectively

dic = {'no': False, 'yes': True}
df['international plan'] = df['international plan'].map(dic)
df.head(3)
```

```python
# Alternatively thi can be achieved using replace()
#Lets use voice mail plan col
dr =df.replace({'voice mail plan': dic})
dr.head(9)
```

```python
#Grouping concept using groupby()
# Combining column --> in this case want to know why one is a churn user based on other features
# Was churn user using: --> day minutes, evening minutes and night minutes

features_col = ['total day minutes', 'total eve minutes', 'total night minutes']
df.groupby(['churn'])[features_col].max()
```

```python
#Info based on categorisation
df.groupby(['churn'])[features_col].describe()
```
